```python
import os
import requests
import time
import random
```

```python
cars = []
with open("carsshort.txt") as fp:
    for line in fp:
        cars.append((int(line.strip().split('|')[0]), line.strip().split('|')[1]))

tracks = []
with open("tracksshort.txt") as fp:
    for line in fp:
        tracks.append((int(line.strip().split('|')[0]),line.strip().split('|')[1]))
```

```python
url = "http://www.radiators-champ.com/RSRLiveTiming/index.php?page=rank&track={0}&car={1}"
headers = {'User-Agent': ('Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/605.1.15'
                          ' (KHTML, like Gecko) Version/14.0 Safari/605.1.15')}
trackcounter = 0
for track, trackname in tracks:
    os.makedirs('rsr_stats/{0}'.format(track), exist_ok=True)
    print('################### starting {0} of {1}'.format(trackcounter + 1, len(tracks)))
    trackcounter += 1
    print("working on track {0} ({1})".format(trackname, track))
    for car, carname in cars:
        if os.path.isfile("rsr_stats/{0}/{1}.html".format(track, car)):
            print("{0} done.".format(carname), )
            continue
        response = requests.get(url.format(track, car), headers=headers)
        if response.status_code == 200:
            with open('rsr_stats/{0}/{1}.html'.format(track, car), 'wb') as fphtml:
                fphtml.write(response.content)
            print("saved car {0} ({1}).".format(carname, car))
        else:
            print("ERROR {2}: car {0} ({1})".format(response.status_code, carname, car))
        time.sleep(2 + random.randrange(3,13))
print("ALL DONE.")
```
